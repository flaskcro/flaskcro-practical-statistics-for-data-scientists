## Bootstrap

1. 샘플 값을 하나 뽑아서 기록하고 제자리에 놓는다.
2. n번 반복한다.
3. 재표본추출된 값의 평균을 기록한다.
4. 1~3단계를 R번 반복한다.
5. R개의 결과를 사용하여
    - a. 그것들의 표준편차를 계산한다.
    - b. 히스토그램 또는 박스 플랏을 그린다.
    - c. 신뢰구간을 찾는다. 

```python
def bootstrap(x, n=30, r=10000):
    return np.array([np.random.choice(x, n, replace=True).mean() for i in range(r)])
```

```python
x = np.random.randint(0,500,100)
```

```python
print(f'mean of x is {x.mean()}, std of x is {x.std()}')
```

```python
print(f'bootstrap mean of x is {bootstrap(x).mean()}, std of x is {bootstrap(x).std()}')
```

```python
bootx = bootstrap(income.x)
```

```python
print(f'mean of income is {income.x.mean()}, std of x is {income.x.std()}')
```

```python
print(f'std / (30 ** 0.5) is {income.x.mean()}, std of x is {income.x.std()/ (30 ** 0.5)}')
```

```python
print(f'mean of income is {bootx.mean()}, std of x is {bootx.std()}')
```

```python
plt.hist(income.x, bins=40)
plt.show()
```

```python
plt.hist(bootx, bins=40)
plt.show()
```

### 표준정규분포와 QQ그림

```python
norm_samp = np.random.normal(0,2, 1000)
```

```python
sns.distplot(norm_samp,bins=30)
mean = norm_samp.mean()
std = norm_samp.std()
std1_lower = mean - std
std1_high = mean + std

std2_lower = mean - 2*std
std2_high = mean + 2*std

std3_lower = mean - 3*std
std3_high = mean + 3*std
plt.axvline(mean, color='red', label='mean')
plt.axvline(std1_lower, color='red', label='mean - std')
plt.axvline(std1_high, color='red', label='mean + std')
plt.axvline(std2_lower, color='navy', label='mean - 2std')
plt.axvline(std2_high, color='navy', label='mean + 2std')
plt.axvline(std3_lower, color='green', label='mean - 3std')
plt.axvline(std3_high, color='green', label='mean + 3std')
plt.show()
```

```python
import statsmodels.api as sm
plt.figure(figsize=(8,6))
sm.qqplot(norm_samp)
mins = norm_samp.min()
maxs = norm_samp.max()
slope = norm_samp.std()
plt.plot([mins, maxs],[mins*slope, maxs*slope], color='red')
plt.show()
```

```python
z = (norm_samp - norm_samp.mean()) / norm_samp.std() 
```

```python
plt.hist(z)
```

```python
len(z[(z < 3) & (z >-3)]) / len(z) * 100.0
```

```python
def z_score(x):
    mean = x.mean()
    std = x.std()
    return [x - mean] / std
```

```python
def drawQQ(x):
    import statsmodels.api as sm
    plt.figure(figsize=(8,6))
    sm.qqplot(x)
    mins = -3
    maxs = 3
    slope = 1
    plt.plot([mins, maxs],[mins*slope, maxs*slope], color='red')
    plt.show()
```

```python
sp500_px = pd.read_csv('./data/sp500_px.csv')
nflx = sp500_px['NFLX'].values
nflx = np.log(nflx[nflx>0])
drawQQ(nflx)
```

```python
nflx = np.diff(nflx)
drawQQ(nflx)
```

```python
drawQQ(nflx[:10])
```

## 푸아송 분포

```python
#분당 2건씩 오는 고객서비스센터가 100분간 받을 문의전화 횟수
s = np.random.poisson(5, 100)
plt.hist(s,bins=10)
plt.show()
```

```python
from scipy.stats import poisson
pdp2 = []
pdp4 = []
pdp10 = []
for num in range(1,40):
    pdp2.append(poisson.pmf(num, 2))
    pdp4.append(poisson.pmf(num, 4))
    pdp10.append(poisson.pmf(num, 10))
plt.plot(pdp2, label=r'$\lambda = $2')
plt.plot(pdp4, label=r'$\lambda = $4')
plt.plot(pdp10, label=r'$\lambda = $10')
plt.legend()
plt.show()
```

```python
pdp2 = []
pdp4 = []
pdp10 = []
for num in range(1,40):
    pdp2.append(poisson.cdf(num, 2))
    pdp4.append(poisson.cdf(num, 4))
    pdp10.append(poisson.cdf(num, 10))
plt.plot(pdp2, label=r'$\lambda = $2')
plt.plot(pdp4, label=r'$\lambda = $4')
plt.plot(pdp10, label=r'$\lambda = $10')
plt.legend()
plt.show()
```
